#**Bell States**

Why it is important:
1. **Foundation for Quantum Entanglement**

2. **Quantum Energy Teleportation:** For quantum teleportation protocols.

3. **Quantum Cryptography:** The properties of Bell states, especially their sensitivity to measurements, underpin quantum key distribution (QKD) protocols such as BB84 and E91. These protocols use entangled states to ensure secure communication, allowing two parties to detect any eavesdropping attempt, which would inevitably alter the state of the entangled qubits. This leads to fundamentally secure communication channels based on the laws of quantum mechanics.

4. **Quantum Computing Algorithms:** For example, for quantum search (Grover's algorithm) and quantum factorization (Shor's algorithm), offering exponential speed-ups for specific computational problems.

## $$\Phi^+ or |00⟩$$  bells state

image.png

image.png

```python
qcBell1 = QuantumCircuit(2) #2 qbit quantum circuit
qcBell1.h(0) #apply hadamard gate on Zeor-th qbit which take the qbit to superposition
qcBell1.cx(0,1) #apply Controled NOT gate on 0 th(control) and 1st (target)
'''#CNOT is a multi qbit gate, takes at least 2 qbits unlike hadamard gate
#first one is control qbit, second one is target qbit. Control qbit is a condition, if it is true, only then target qbit is changed and executed, otherwise not
#the condition is that if there is 1 in the control qbit, it will operate on the target qbit and make it 1 from 0'''
qcBell1.measure_all() #apply measurement on both qbits
```

```python
#select simulator from Aer package
simulator = Aer.get_backend('aer_simulator')

qcBell1.save_statevector() #statevector = state + vector -> state of the qbit in a vector format

circuit1 = transpile(qcBell1, simulator) #Transpile() converts quntumcircuits to sorcecode for the simulator

# Run and get statevector
result = simulator.run(circuit1,shots = 500).result()
statevector = result.get_statevector(circuit1)
print(result.get_counts())
qcBell1.draw('mpl')
```

```python
print(statevector) #Printing state vector
plot_histogram(result.get_counts(circuit1))
```
